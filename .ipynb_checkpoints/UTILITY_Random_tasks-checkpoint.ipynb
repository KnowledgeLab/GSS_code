```python
import numpy as np
import pandas as pd 
import cPickle
```

```python
# convert updated (1972-2016) GSS dataset to cPickle format
# dataset is in Stata form, downloaded from http://gss.norc.org/get-the-data/stata on 2018-03-26

pathToDf = '../Data/GSS Dataset/stata/'
df = pd.read_stata(pathToDf + 'GSS7216_R3.DTA', convert_categoricals=False)   #note, convert_categoricals=True doesn't even work, and throws an error: https://github.com/pandas-dev/pandas/issues/13923
df.index = df['year']
df.columns = map(str.upper, df.columns)

# for comparison
# df2 = pd.read_pickle('../Data/df.pickle')
```

```python
df.head()
```

```python
cPickle.dump(df, open('../Data/GSS Dataset/stata/GSS7216_R3.cpickle', 'wb'))
```

# Create VARS_BY_YEAR.pickle

Needs the following form: type==dict, where key=year, value=set('VAR1', 'VAR2', ...)

The previous file I used to use (and is now the backup) is: ../Data/VARS_BY_YEAR_1972-2012.pickle


```python
VARS_BY_YEAR_old = cPickle.load(open('../Data/VARS_BY_YEAR_1972-2012.pickle', 'rb'))
df = cPickle.load(open('../Data/GSS Dataset/stata/GSS7216_R3.cpickle', 'rb'))
```

```python
# for some reason, the VARS_BY_YEAR_old is a bit smaller (~66 variables) than the new combined dataset (df).
# and VARS_BY_YEAR_old also has ~2 variables that the newer one doesn't have. So I'll add those on too

for year in df.index.unique(): # 1972, 1973, ...
    print 'processing', year
    valid_variables = df.loc[year].columns[df.loc[year].notnull().any(axis=0)]
    valid_variables = set(map(str, valid_variables))
    if year<2014:
        valid_variables.update(VARS_BY_YEAR_old[year])
    VARS_BY_YEAR[year] = valid_variables
```

```python
cPickle.dump(VARS_BY_YEAR, open('../Data/VARS_BY_YEAR.cpickle', 'wb'))
```
