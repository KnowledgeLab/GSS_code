```python
import numpy as np
import pandas as pd 
import cPickle
```

```python
# convert updated (1972-2016) GSS dataset to cPickle format
# dataset is in Stata form, downloaded from http://gss.norc.org/get-the-data/stata on 2018-03-26

pathToDf = '../Data/GSS Dataset/stata/'
df = pd.read_stata(pathToDf + 'GSS7216_R3.DTA', convert_categoricals=False)   #note, convert_categoricals=True doesn't even work, and throws an error: https://github.com/pandas-dev/pandas/issues/13923
df.index = df['year']
df.columns = map(str.upper, df.columns)

# for comparison
# df2 = pd.read_pickle('../Data/df.pickle')
```

```python
cPickle.dump(df, open('../Data/GSS Dataset/stata/GSS7216_R3.cpickle', 'wb'))
```
